```python
#引用的第三方包
%matplotlib inline
import codecs
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import matplotlib.pyplot as plt
import pickle
import jieba
from os import path
from PIL import Image
import numpy as np
import random
```

```python
#文件路径
stopwords=r'.\data\stopwords.txt'
e_text_file_path=r'.\data\What_I_Have_Lived_For.txt'#ryfs英文文件
c_text_file_path=r'.\data\santi.txt'#ryfs中文文件
mask_path=r'.\data\apple.jpg'#遮罩路径
font_path=r'C:/Users/Windows/fonts/msyh.ttf'#字体路径
```

## wordcloud

```python
#预处理

#英文暂时不需要预处理
etext=open(e_text_file_path,encoding='utf-8').read()

#中文预处理
#分词
ctext=open(c_text_file_path,encoding='utf-8').read()
#seg_list=list(jieba.cut(ctext,cut_all=False))
cut_ctext=" ".join(jieba.cut(ctext,cut_all=False))

#指定遮罩
phone_mask=np.array(Image.open(mask_path))#词云形状的来源图片
```

```python
#定义词云
wc=WordCloud(
    font_path=font_path,#设置字体格式,如不设置显示不了中文
    width=1000,#词云画布尺寸
    height=1000,#词云画布尺寸
    scale=1,
    background_color="black",
    max_words=100,
    mask=phone_mask,#mask遮罩，数组
    stopwords=STOPWORDS.add("said"))
#参数说明
#font_path='C:/Users/Windows/fonts/msyh.ttf',#设置字体格式,如不设置显示不了中文
#width=900,height=500,词云画布尺寸
#prefer_horizontal,是否接受选择，未测试
#mask遮罩，数组
#scale尺度
#min_font_size，max_font_size最小大字号
#font_step
#max_words
#stopwords
#background_color    black，white
#mode
#normalize_plurals
```

```python
wc.generate(etext)#text,ctext,etext
#绘图,英文词云图
plt.imshow(wc)
plt.axis("off")
#保存到文件
wc.to_file('etext.png')
```

```python
wc.generate(cut_ctext)
#绘图
plt.imshow(wc)
plt.axis("off")
#保存到文件
wc.to_file('ctext.png')
```
